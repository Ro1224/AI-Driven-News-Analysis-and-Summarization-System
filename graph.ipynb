```python
import matplotlib.pyplot as plt

# Languages and ROUGE-L scores
languages = ['English', 'French', 'Hindi', 'Chinese', 'Thai', 'Turkish', 'Tamil']
mbart_scores = [34.9, 30.3, 24.5, 32.7, 34.0, 3.0, 78.2]
mt5_scores   = [34.2, 29.4, 23.5, 32.1, 33.4, 3.0, 72.5]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(languages, mbart_scores, marker='o', label='mBART', color='royalblue')
plt.plot(languages, mt5_scores, marker='s', label='mT5', color='red')

# Adding annotations
for i, lang in enumerate(languages):
    plt.text(lang, mbart_scores[i] + 1, f'{mbart_scores[i]:.1f}', ha='center', fontsize=8)
    plt.text(lang, mt5_scores[i] - 4, f'{mt5_scores[i]:.1f}', ha='center', fontsize=8)

# Labels and formatting
plt.title('Multilingual Model ROUGE-L Score Comparison: mBART vs. mT5')
plt.xlabel('Language')
plt.ylabel('ROUGE-L Score')
plt.ylim(0, 85)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Languages and BLEU scores
languages = ['English', 'French', 'German', 'Spanish', 'Chinese', 'Tamil']
mariam_mt_scores = [45.2, 42.1, 39.8, 43.0, 41.0, 38.5]  # Example scores for MARIAM MT
mt5_scores = [44.0, 40.5, 38.9, 42.3, 39.5, 36.2]  # Example scores for mT5

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(languages, mariam_mt_scores, marker='o', label='MARIAM MT', color='red')
plt.plot(languages, mt5_scores, marker='s', label='mT5', color='green')

# Adding annotations
for i, lang in enumerate(languages):
    plt.text(lang, mariam_mt_scores[i] + 0.5, f'{mariam_mt_scores[i]:.1f}', ha='center', fontsize=8)
    plt.text(lang, mt5_scores[i] - 0.5, f'{mt5_scores[i]:.1f}', ha='center', fontsize=8)

# Labels and formatting
plt.title('Machine Translation Comparison: MARIAM MT vs. mT5 (BLEU Scores)')
plt.xlabel('Language')
plt.ylabel('BLEU Score')
plt.ylim(30, 50)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Benchmarks and ROUGE-L scores for BART and PEGASUS
benchmarks = ['CNN/Daily Mail', 'XSum', 'Gigaword', 'Newsroom', 'New York Times']
bart_scores = [42.5, 46.1, 38.2, 44.3, 45.2]  # Example ROUGE-L scores for BART
pegasus_scores = [40.6, 44.2, 35.5, 42.1, 43.0]  # Example ROUGE-L scores for PEGASUS

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(benchmarks, bart_scores, marker='o', label='BART', color='yellow')
plt.plot(benchmarks, pegasus_scores, marker='s', label='PEGASUS', color='orange')

# Adding annotations
for i, benchmark in enumerate(benchmarks):
    plt.text(benchmark, bart_scores[i] + 0.5, f'{bart_scores[i]:.1f}', ha='center', fontsize=8)
    plt.text(benchmark, pegasus_scores[i] + 0.5, f'{pegasus_scores[i]:.1f}', ha='center', fontsize=8)

# Labels and formatting
plt.title('Summarization Model Comparison: BART vs. PEGASUS (ROUGE-L Scores)')
plt.xlabel('Benchmark Dataset')
plt.ylabel('ROUGE-L Score')
plt.ylim(30, 50)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# News Articles and Credibility Scores for RoBERTa (with and without sentiment analysis)
articles = ['Article 1', 'Article 2', 'Article 3', 'Article 4', 'Article 5']
roberta_with_sentiment = [0.85, 0.78, 0.90, 0.75, 0.80]  # Example credibility scores with sentiment analysis
roberta_without_sentiment = [0.80, 0.74, 0.85, 0.70, 0.76]  # Example credibility scores without sentiment analysis

# Create the plot
x = range(len(articles))  # Article indices for plotting

plt.figure(figsize=(10, 6))

# Plotting the bar chart for both models
bar_width = 0.35
plt.bar(x, roberta_with_sentiment, width=bar_width, label='RoBERTa with Sentiment Analysis', color='royalblue')
plt.bar([p + bar_width for p in x], roberta_without_sentiment, width=bar_width, label='RoBERTa without Sentiment Analysis', color='orange')

# Adding annotations for each bar
for i, value in enumerate(roberta_with_sentiment):
    plt.text(i, value + 0.01, f'{value:.2f}', ha='center', fontsize=10)
for i, value in enumerate(roberta_without_sentiment):
    plt.text(i + bar_width, value + 0.01, f'{value:.2f}', ha='center', fontsize=10)

# Labels and formatting
plt.title('News Credibility Comparison: RoBERTa with vs without Sentiment Analysis')
plt.xlabel('News Articles')
plt.ylabel('Credibility Score')
plt.xticks([p + bar_width / 2 for p in x], articles)  # Position x-ticks in the middle of the bars
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()
```
